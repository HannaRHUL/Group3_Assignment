### 2.4 Merge data from multiple files into a single DataFrame

```python
# (1) Athlete + NOC
df_athlete_and_NOC = pd.merge(df_athlete, df_NOC, on="NOC")

# (2) gdp + income
df_gdp_and_income = pd.merge(
    gdp,
    income_group,
    how="left",
    left_on="Country Code",  # Replace with the exact column name in main_data
    right_on="Country Code"  # Replace with the exact column name in metadata_country
)
df_economic_power = pd.merge(
   df_gdp_and_income,
    info,
    how="left",
    left_on="Indicator Code",  # Replace with the exact column name in main_data
    right_on="INDICATOR_CODE"  # Replace with the exact column name in metadata_indicator
)

#There are two columns for both indicator code and indicator name so we drop the duplicate columns
df_economic_power = df_economic_power.drop(columns=["INDICATOR_CODE", "INDICATOR_NAME"])


# (3) Athlete and NOC + Economic Power
# Before merging, we have to change some column names and remove columns of economic power Data frame

df_economic_power.rename(columns={'Region': 'SubContinent'}, inplace=True)

id_vars = df_economic_power.select_dtypes(exclude='number').columns.tolist()
df_melted_economic_power = df_economic_power.melt(
    id_vars=id_vars,
    var_name='Year',
    value_name='GDP'
)

# (4) Merge df_athlete_and_NOC + df_melted_economic_power
# Handle missing values in the 'Year' columns before converting to integers
# Replace NaN with 0 (or any other default value) and then convert to int
df_athlete_and_NOC['Year'] = pd.to_numeric(df_athlete_and_NOC['Year'], errors='coerce')  # Convert non-numeric to NaN
df_athlete_and_NOC['Year'] = df_athlete_and_NOC['Year'].fillna(0).astype(int)  # Fill NaN with 0 and convert to int

df_melted_economic_power['Year'] = pd.to_numeric(df_melted_economic_power['Year'], errors='coerce')  # Convert non-numeric to NaN
df_melted_economic_power['Year'] = df_melted_economic_power['Year'].fillna(0).astype(int)  # Fill NaN with 0 and convert to int

# Merge the two DataFrames using both 'NOC' and 'Year' as keys
df = pd.merge(
    df_athlete_and_NOC,
    df_melted_economic_power,
    how="inner",  # Perform an inner join to keep only matching rows
    left_on=["NOC", "Year"],  # Use 'NOC' and 'Year' from the left DataFrame
    right_on=["Country Code", "Year"]  # Use 'Country Code' and 'Year' from the right DataFrame
)
```

```python
# (3) Create Century Column
def get_century_suffix(year):
    century = year // 100 + 1
    if century % 10 == 1 and century != 11:
        suffix = 'st'
    elif century % 10 == 2 and century != 12:
        suffix = 'nd'
    elif century % 10 == 3 and century != 13:
        suffix = 'rd'
    else:
        suffix = 'th'
    return f"{century}{suffix}"
df['Century'] = df['Year'].apply(get_century_suffix)
```

### 5.4 Analysis 4 : Sports with highest/lowest BMI

```python
# (4-a) Calculate BMI for each athlete
df['BMI'] = df['Weight'] / (df['Height'] / 100) ** 2


# (4-b) Group by 'Sport' and calculate the average BMI for each sport, rounding to one decimal place
average_bmi_per_sport = df.groupby('Sport')['BMI'].mean().round(1).reset_index()

# (4-c) Sort by average BMI to get the top 5 highest and lowest for male athletes
top_5_highest_bmi = average_bmi_per_sport.sort_values(by='BMI', ascending=False).head(5).reset_index(drop=True)
top_5_lowest_bmi = average_bmi_per_sport.sort_values(by='BMI', ascending=True).head(5).reset_index(drop=True)


# (4-d) Add Rank columns
top_5_highest_bmi['Rank'] = range(1, 6)
top_5_lowest_bmi['Rank'] = range(1, 6)


# (4-e) Display the result
print("Analysis 4 : sports with highest/lowest BMI")
print("Top 5 Sports with the Highest and Lowest Average BMI:")
for high, low in zip(top_5_highest_bmi.itertuples(index=False), top_5_lowest_bmi.itertuples(index=False)):
    print(f"Rank {high.Rank}: {high.Sport} (BMI: {high.BMI})".ljust(50) + f"Rank {low.Rank}: {low.Sport} (BMI: {low.BMI})")
print("\n\n")
```

### 5.6 Analysis 6: Relationship between age and average number of gold medals

```python
# (6-a) Define correlation calculation functions
def pearson_correlation(x, y):
    """
    Calculate Pearson's correlation coefficient between two lists.
    
    Args:
        x (list): First variable data.
        y (list): Second variable data.
        
    Returns:
        float: Pearson's correlation coefficient.
    """
    mean_x = sum(x) / len(x)
    mean_y = sum(y) / len(y)
    numerator = sum((xi - mean_x) * (yi - mean_y) for xi, yi in zip(x, y))
    denominator = (
        (sum((xi - mean_x) ** 2 for xi in x) * sum((yi - mean_y) ** 2 for yi in y)) ** 0.5
    )
    return numerator / denominator if denominator != 0 else 0

def spearman_correlation(x, y):
    """
    Calculate Spearman's rank correlation coefficient between two lists.
    This is achieved by ranking the data and applying Pearson's formula to the ranks.
    
    Args:
        x (list): First variable data.
        y (list): Second variable data.
        
    Returns:
        float: Spearman's rank correlation coefficient.
    """
    def rank(data):
        unique_data = sorted(set(data))
        rank_map = {value: rank for rank, value in enumerate(unique_data, 1)}
        return [rank_map[val] for val in data]
    
    x_ranks = rank(x)
    y_ranks = rank(y)
    return pearson_correlation(x_ranks, y_ranks)

# Use existing DataFrame with 'Age' and 'Medal' columns
df_age_medal = df 

# (6-b) Group the data by 'Age' to calculate the total number of medals for each age
# Create a binary column to indicate if a medal was awarded
df_age_medal['Medal_Binary'] = df_age_medal['Medal'].apply(
    lambda x: 1 if x in ['Gold', 'Silver', 'Bronze'] else 0
)
# Determine the dynamic age range from the dataset
min_age = int(df_age_medal['Age'].min())
max_age = int(df_age_medal['Age'].max())
print(f"Dynamic age range in the dataset: {min_age} to {max_age}")

# Group by age and calculate the total medals
age_medal_counts = df_age_medal.groupby('Age')['Medal_Binary'].sum().reset_index()

# Ensure all ages in the dynamic range are included
all_ages = pd.DataFrame({'Age': np.arange(min_age, max_age + 1)})
age_medal_counts = all_ages.merge(age_medal_counts, on='Age', how='left').fillna(0)

# Ensure the Medal_Binary column is integer after filling NaNs
age_medal_counts['Medal_Binary'] = age_medal_counts['Medal_Binary'].astype(int)

# (6-c) Compute Spearman's Rank Correlation
correlation = spearman_correlation(age_medal_counts['Age'], age_medal_counts['Medal_Binary'])

# Display the result
print("Analysis 6: Relationship between age and average number of gold medals")
print(f"Spearman Correlation Coefficient: {correlation:.2f}")
```

### 5.7 Analysis 7: GDP and Medal Binary Relationship

```python
# Use existing DataFrame with 'GDP' and 'Medal' columns
df_gdp_medal = df.copy()

# Ensure GDP and Medal_Binary are numeric
df_gdp_medal['GDP'] = pd.to_numeric(df_gdp_medal['GDP'], errors='coerce')
df_gdp_medal['Medal_Binary'] = df_gdp_medal['Medal'].apply(
    lambda x: 1 if x in ['Gold', 'Silver', 'Bronze'] else 0
)

# Drop rows with missing values in either 'GDP' or 'Medal_Binary'
df_gdp_medal = df_gdp_medal.dropna(subset=['GDP', 'Medal_Binary'])

# Aggregate medals by GDP
gdp_medal_aggregated = df_gdp_medal.groupby('GDP')['Medal_Binary'].sum().reset_index()

# Ensure enough data points for Pearson's correlation
if len(gdp_medal_aggregated) < 2:
    print("Not enough valid data points to compute Pearson's correlation.")
else:
    # Extract GDP and Medal counts
    gdp_values = gdp_medal_aggregated['GDP']
    medal_counts = gdp_medal_aggregated['Medal_Binary']

    # Compute Pearson's correlation
    correlation = pearson_correlation(gdp_values, medal_counts)

    # Print results
    print("Analysis 7: GDP and Medal Binary Relationship")
    print(f"Pearson Correlation Coefficient: {correlation:.2f}")
    if abs(correlation) > 0.5:  # Basic significance threshold
        print("There is a significant linear relationship between GDP and medals won.")
    else:
        print("There is NO significant linear relationship between GDP and medals won.")
```

### 6.3 Visualisation 3: Correlation between GDP and number of Medals Won

```python
# Function to create the interactive plot for top 20 countries
def create_interactive_plot(df, medal_type="Gold"):
    # Filter the data for the selected medal type
    if medal_type == "Gold":
        medal_count = df[df["Medal"] == "Gold"].groupby("Team").size().reset_index(name="Medals")
    else:  # Total medals
        medal_count = df.groupby("Team").size().reset_index(name="Medals")

    # Merge with GDP data
    gdp_data = df[["Team", "GDP"]].drop_duplicates(subset=["Team"])
    merged_data = pd.merge(medal_count, gdp_data, on="Team")

    # Add GDP in trillions for readability
    merged_data["GDP (in Trillions USD)"] = merged_data["GDP"] / 1e12

    # Select top 20 countries based on medals
    top_countries = merged_data.sort_values(by="Medals", ascending=False).head(20)

   # Create the interactive plot
    scatter_plot = top_countries.hvplot.scatter(
    x="GDP (in Trillions USD)",
    y="Medals",
    size=top_countries["Medals"] / 10,  # Reduce size by dividing medals by 10
    hover_cols=["Team", "GDP (in Trillions USD)", "Medals"],  # Include explanatory columns in hover
    color="Medals",  # Color represents the number of medals
    cmap="coolwarm",
    title=f"Olympics: Top 20 Teams by {medal_type} Medals and GDP",
    height=500,
    width=800,
)



    # Add descriptive text for size and colour meanings
    scatter_plot = scatter_plot.opts(
        tools=["hover"],
        xlabel="GDP (in Trillions USD)",
        ylabel=f"Number of {medal_type} Medals",
        fontsize={"title": 14, "labels": 12, "ticks": 10},
    )

    return scatter_plot

# Dropdown to toggle between medal types
# Create a dropdown for medal type
medal_selector = pn.widgets.Select(name="Medal Type", options=["Gold", "Total"], value="Gold")

# Dynamic plot based on the selected medal type
@pn.depends(medal_selector.param.value)
def update_plot(medal_type):
    return create_interactive_plot(df, medal_type)

# Add a description to the dashboard
description = """
### Olympics: Medal and GDP Visualisation
- **Size of Points**: Represents the number of medals won by the country.
- **Color of Points**: Reflects the number of medals won (darker = more medals).
"""

# Layout with dropdown, description, and plot
dashboard = pn.Column(
    pn.pane.Markdown(description),  # Add the description at the top
    medal_selector,
    update_plot,
)
dashboard.servable()
```
